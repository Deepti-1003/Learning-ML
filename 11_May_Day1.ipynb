Training a data model set for function f(x)=3x+1

```python
#importing all libraries necessary for the program
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
#creating a neural network, it has 1 layer and that layer has 1 neuron
model=tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```python
#compiling the neural network with optimier and loss
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
#xs is input dataset and ys is output dataset
xs=np.array([1.0,2.0,3.0,4.0],dtype=float)
ys=np.array([4.0,7.0,10.0,13.0],dtype=float)
```

```python
#Training the neural network where epochs is number of loop set to train data
model.fit(xs,ys,epochs=500)
```

```python
#Printing the prediction for 10
td=[10.0,10.5,11]
tp=model.predict(td)
print(tp)
```

```python
plt.plot(xs, ys, 'ro', label ='Original data')
plt.plot(td,tp,'bo',label='predicted data')
plt.title('Linear Regression Result')
plt.legend()
plt.show()
```
